```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
plt.figure(figsize=(12, 8))
sns.scatterplot(x='PriceEuro', y='Model', hue='Model', data=df, palette='tab20', s=100)
plt.title('Price vs. Model')
plt.xlabel('Price (Euro)')
plt.ylabel('Model')
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
# Select relevant columns
data_subset = df[['TopSpeed_KmH', 'Brand', 'Model']]

# Create a pivot table using mean as an aggregation function
pivot_data = data_subset.pivot_table(index='Brand', columns='Model', values='TopSpeed_KmH', aggfunc='mean')

# Plotting the grouped bar plot
plt.figure(figsize=(14, 8))
pivot_data.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Top Speed vs. Brand and Model')
plt.xlabel('Brand')
plt.ylabel('Top Speed (Km/H)')
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Assuming your DataFrame is named 'df'
df_encoded = df.copy()

# Identify categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Apply label encoding for categorical columns
label_encoder = LabelEncoder()
for column in categorical_columns:
    df_encoded[column] = label_encoder.fit_transform(df[column])

# Display the encoded DataFrame
print(df_encoded)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'df_encoded'
# Selecting relevant columns
X = df_encoded[['PriceEuro', 'Model']]  # Assuming 'Model_encoded' is the encoded version of 'Model'

# Defining the number of clusters (you can adjust this based on your preference)
n_clusters = 5

# Creating and fitting the KMeans model
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Adding cluster labels to the original DataFrame
df_encoded['Cluster'] = y_kmeans

# Plotting the clusters
plt.figure(figsize=(15, 7))

# Scatter plot for each cluster
for cluster_label in range(n_clusters):
    sns.scatterplot(x=df_encoded[df_encoded['Cluster'] == cluster_label]['PriceEuro'],
                    y=df_encoded[df_encoded['Cluster'] == cluster_label]['Model'],
                    label=f'Cluster {cluster_label + 1}', s=50)

# Add cluster centers
cluster_centers = kmeans.cluster_centers_
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='black', marker='x', s=100, label='Centroids')

plt.title('K-means Clustering based on Price and Model')
plt.xlabel('Price (Euro)')
plt.ylabel('Model')
plt.legend()
plt.show()
```
